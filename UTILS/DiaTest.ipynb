```python
import mermaid as md
from mermaid.graph import Graph
```

```python
graphe_2 = Graph('class diagram 1',"""
classDiagram
    Animal <|-- Duck
    Animal <|-- Fish
    Animal <|-- Zebra
    Animal : +int age
    Animal : +String gender
    Animal: +isMammal()
    Animal: +mate()
    class Duck{
      +String beakColor
      +swim()
      +quack()
    }
    class Fish{
      -int sizeInFeet
      -canEat()
    }
    class Zebra{
      +bool is_wild
      +run()
    }
""")
```

```python
md.Mermaid(graphe_2)
```
