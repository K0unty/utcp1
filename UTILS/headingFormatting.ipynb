# Heading Formatting 

> Testing out the formatting of the heading in this nb 

## Adding Simple Background Color and foreground Color

Eg 
<h1 style="background-color:#020617; color:#22c55e;padding:40px; text-align:center">My Heading</h1>


> Code for above

```sh 
<h1 style="background-color:#020617; color:white;padding:40px; text-align:center">My Heading</h1>
```
## Adding a nice BG block 

<div style="background: url(https://i.ibb.co/JdDDCYZ/image.png) repeat; color: #00ff00; padding: 2px; text-align: center">
    <h1>My Heading</h1>
</div>

Code for above
```sh 
<div style="background: url(https://i.ibb.co/JdDDCYZ/image.png) repeat; color: #00ff00; padding: 2px; text-align: center">
    <h1>My Heading</h1>
</div>

```
