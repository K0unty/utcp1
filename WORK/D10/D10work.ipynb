1. Function with outputs - This is the main type of function which you really need 

```python
# Function capital casing or title casing

def format_name(f_name, l_name):
    '''
    Function to format the name by capitalzing
    '''
    f_n = f_name.title()
    l_n = l_name.title()
    return f'{f_n} {l_n}'
    
# Execute above function 
format_name("booty", "queen")
format_name("booty", "qUeEn")
```

```python
# Having many return function 
def format_name(f_name, l_name):
    '''
    Function to format the name by capitalzing
    '''
    if f_name == "" or l_name == "":
        return "Nothing Entered"
    f_n = f_name.title()
    l_n = l_name.title()
    return f'{f_n} {l_n}'

rprint(format_name("", ""))
```

## 103 - ICE - DaysOfMonth <a class="anchor" id="s103"></a> [↩️TOC](#toc)

Make program days in a month and year

```python
# finding out if the year give is a leap year 

def is_leap(year):
    if year % 4 == 0:
        if year % 100 == 0 :
            if year % 400 == 0:
                rprint(f'[#afffaf] LEAP')
            else:
                rprint(f'[#afffff] NOLEAP')
        else: 
            rprint(f'[#afffaf] LEAP')
    else:
        rprint(f'[#afffff] NOLEAP')
        
# Check function 
is_leap(2021)
```

```python
### Mkaing changes as per excercise 
def is_leap(year):
    if year % 4 == 0:
        if year % 100 == 0 :
            if year % 400 == 0:
                return True
            else:
                return False
        else: 
            return True
    else:
        return False
        
# Second fuction 
def days_in_month(year, month):
    month_days = [31,28,31,30,31,30,31,31,30, 31,30,31]
    if month == 2 and is_leap(year):
        return 29
    else:
        return month_days[month - 1]

# given Code 
year = int(input()) # Enter Year 
month = int(input()) # Enter month 
days = days_in_month(year, month)
rprint(days)
```
