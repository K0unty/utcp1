```python
# Function capital casing or title casing

def format_name(f_name, l_name):
    '''
    Function to format the name by capitalzing
    '''
    f_n = f_name.title()
    l_n = l_name.title()
    return f'{f_n} {l_n}'
    
# Execute above function 
format_name("booty", "queen")
format_name("booty", "qUeEn")
```

## 103 - ICE - DaysOfMonth <a class="anchor" id="s103"></a> [↩️TOC](#toc)

Make program days in a month and year

```python
# finding out if the year give is a leap year 

def is_leap(year):
    if year % 4 == 0:
        if year % 100 == 0 :
            if year % 400 == 0:
                rprint(f'[#afffaf] LEAP')
            else:
                rprint(f'[#afffff] NOLEAP')
        else: 
            rprint(f'[#afffaf] LEAP')
    else:
        rprint(f'[#afffff] NOLEAP')
        
# Check function 
is_leap(2021)
```

```python
### Mkaing changes as per excercise 
def is_leap(year):
    if year % 4 == 0:
        if year % 100 == 0 :
            if year % 400 == 0:
                return True
            else:
                return False
        else: 
            return True
    else:
        return False
        
# Second fuction 
def days_in_month(year, month):
    month_days = [31,28,31,30,31,30,31,31,30, 31,30,31]
    if month == 2 and is_leap(year):
        return 29
    else:
        return month_days[month - 1]

# given Code 
year = int(input()) # Enter Year 
month = int(input()) # Enter month 
days = days_in_month(year, month)
rprint(days)
```

## 105 - Calculator - Part1: Combining Dictionaries and Functions <a class="anchor" id="s105"></a> [↩️TOC](#toc)

```python
# Basic caclulator functions 

#add 
def add(n1,n2):
    return n1 + n2

# subtract
def subtract(n1,n2):
    return n1 - n2

# Multiply
def multiply(n1,n2):
    return n1 * n2

# Divide
def divide(n1,n2):
    return n1 / n2

# Makind dictionary  - Note this dictionary values are the name of the funtions defind above

operations = {
    "+": add, 
    "-": subtract,
    "*" : multiply,
    "/": divide,
}
num1 = int(input("Enter 1"))

for symbol in operations:
    rprint(symbol)

operation_symbol = input("Pic OP")
num2 = int(input("Enter 2"))

calculation_function = operations[operation_symbol]
answer = calculation_function(num1, num2)

rprint(f'{num1} {operation_symbol} {num2} = {answer}')
```

```python
# Basic caclulator functions 

#add 
def add(n1,n2):
    return n1 + n2

# subtract
def subtract(n1,n2):
    return n1 - n2

# Multiply
def multiply(n1,n2):
    return n1 * n2

# Divide
def divide(n1,n2):
    return n1 / n2

# Makind dictionary  - Note this dictionary values are the name of the funtions defind above

operations = {
    "+": add, 
    "-": subtract,
    "*" : multiply,
    "/": divide,
}
num1 = int(input("Enter 1"))

# Printing the functions
for symbol in operations:
    rprint(symbol)
    
# Variable used for exiting the loop
should_continue = True

while should_continue == True:
    # get input of the symbol
    operation_symbol = input("Pic OP")

    # Get the second number
    num2 = int(input("Next Number: "))

    # 1. Choose the corresponding function based on the symbol
    calculation_function = operations[operation_symbol]

    # 2. Calling the operation above , in calculation_function , and the using that as the functon to do operation of the inputs
    answer = calculation_function(num1, num2)

    rprint(f'{num1} {operation_symbol} {num2} = {answer}')

    input(f'y for calculating with {answer}')
    
    if input(f'y for calculating with {answer}, or n to exit') == "y":
        num1 = answer 
    else: 
        should_continue = False
        
```
