```python
import random
from typing import List, Tuple

# Constants
ACE_VALUE = 11
FACE_CARD_VALUE = 10
BLACKJACK_VALUE = 21
DEALER_STAND_VALUE = 17

CARD_VALUES = [ACE_VALUE, 2, 3, 4, 5, 6, 7, 8, 9, FACE_CARD_VALUE, FACE_CARD_VALUE, FACE_CARD_VALUE, FACE_CARD_VALUE]

def deal_card() -> int:
    """
    Returns a random card from the deck.
    
    Returns:
        int: The value of the dealt card.
    """
    return random.choice(CARD_VALUES)

def calculate_score(cards: List[int]) -> int:
    """
    Calculate the score of a hand, accounting for Blackjack and Ace values.
    
    Args:
        cards (List[int]): A list of card values.
    
    Returns:
        int: The calculated score. Returns 0 for a Blackjack.
    """
    if sum(cards) == BLACKJACK_VALUE and len(cards) == 2:
        return 0  # Represents a Blackjack

    if ACE_VALUE in cards and sum(cards) > BLACKJACK_VALUE:
        cards = [1 if card == ACE_VALUE else card for card in cards]

    return sum(cards)

def compare_scores(user_score: int, dealer_score: int) -> str:
    """
    Compare the scores and determine the game outcome.
    
    Args:
        user_score (int): The user's score.
        dealer_score (int): The dealer's score.
    
    Returns:
        str: A message indicating the game outcome.
    """
    if user_score == dealer_score:
        return "It's a draw!"
    elif dealer_score == 0:
        return "You lose. Dealer has Blackjack."
    elif user_score == 0:
        return "You win with a Blackjack!"
    elif user_score > BLACKJACK_VALUE:
        return "You went over. You lose."
    elif dealer_score > BLACKJACK_VALUE:
        return "Dealer went over. You win!"
    elif user_score > dealer_score:
        return "You win!"
    else:
        return "You lose."

def play_game() -> None:
    """
    Play a single game of Blackjack.
    """
    user_hand = [deal_card() for _ in range(2)]
    dealer_hand = [deal_card() for _ in range(2)]

    game_over = False
    while not game_over:
        user_score = calculate_score(user_hand)
        dealer_score = calculate_score(dealer_hand)

        print(f"Your cards: {user_hand}, current score: {user_score}")
        print(f"Dealer's first card: {dealer_hand[0]}")

        if user_score == 0 or dealer_score == 0 or user_score > BLACKJACK_VALUE:
            game_over = True
        else:
            should_continue = input("Type 'y' to get another card, 'n' to pass: ").lower()
            if should_continue == 'y':
                user_hand.append(deal_card())
            else:
                game_over = True

    while dealer_score != 0 and dealer_score < DEALER_STAND_VALUE:
        dealer_hand.append(deal_card())
        dealer_score = calculate_score(dealer_hand)

    print(f"Your final hand: {user_hand}, final score: {user_score}")
    print(f"Dealer's final hand: {dealer_hand}, final score: {dealer_score}")
    print(compare_scores(user_score, dealer_score))

def main() -> None:
    """
    Main game loop.
    """
    while input("Do you want to play a game of Blackjack? Type 'y' or 'n': ").lower() == 'y':
        play_game()

if __name__ == "__main__":
    main()
```
