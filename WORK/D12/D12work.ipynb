# 120 - How to Modify a Global Variable <a class="anchor" id="s120"></a> [↩️TOC](#toc)

```python
# Modifying the global scope

ene = 1

def inc_ene():
    
    # Changing the global var
    global ene
    
    ene += 1
    rprint(f'ene inside func: {ene}')

inc_ene()
rprint(f'ene outside func: {ene}')
```

1. Using return statements

```python
# Modifying the global scope

ene2 = 1

def inc_ene2():
    return ene2 + 100

ene2 = inc_ene2()
rprint(f'ene outside func: {ene2}')
```

# 123 - The Number Guessing Game <a class="anchor" id="s123"></a> [↩️TOC](#toc)

1. Advice on doing this project

```python
banner = '''
**********************************************************
GUESS PUSSY SMELL - GUESS PUSSY SMELL - GUESS PUSSY SMELL
GUESS PUSSY SMELL - GUESS PUSSY SMELL - GUESS PUSSY SMELL
GUESS PUSSY SMELL - GUESS PUSSY SMELL - GUESS PUSSY SMELL
***********************************************************
'''

# Choosing a random number 1 to 100 

from random import randint as rp

# Global Constant for choosing level difficulty 
EA_LE = 10 
HA_LE = 5


# make function to compare guessed number for chosen number 
def check_answer(guess, answer):
    if guess > answer:
        rprint(f'[#ff0000]😡 TOO HIGH')
    elif guess < answer :
        rprint(f'[#ff00d7]😡 TOO LOW')
    else: 
        rprint(f'[#d7ff00]👄 StinkIsRight {answer}')

        
# Make Function to set a difficulty 
def set_difficulty():
    level = Prompt.ask('Easy(E) or Hard(H) Fukr ?:', choices = ["E", "H"])
    if level == "E":
        return EA_LE
    else:
        return HA_LE
        

#  Make game function 

def game():
    rprint(f'''
    [#ff5f00]{banner}
    Smell her stink!!!

    ''')
    rprint(f'''Guess strenght of pussy stink from 1 to 100''')
    answer = rp(1,100)
    rprint(f'Expected - {answer}')

    turns = set_difficulty()

    guess = 0
    while guess != answer:
        rprint(f'You have {turns} Attempts')
        guess = int(Prompt.ask("Guess Number: "))
        check_answer(guess, answer)

# Call game function 
game()
```
