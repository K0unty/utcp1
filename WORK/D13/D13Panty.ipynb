```python
from rich.traceback import install

def mutate(a_list):
    b_list = []
    for i in a_list:
        new_item = i * 2
        b_list.append(new_item)
    print(b_list)

# Install rich traceback handler
install(show_locals=True)

# Execute function
mutate([1,2,3,5,8,13])
```
