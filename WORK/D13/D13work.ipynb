````python
# Problematic code 
def mutate(a_list):
    b_list = []
    for i in a_list:
        new_item = i * 2
        rprint(new_item)
    b_list.append(new_item)
    rprint(b_list)

# Execute function
mutate([1,2,3,5,8,13])

````

1. To use this debugger you have to use the following
2. `p` to print any specific variable at that point 
3. `n` execute the next line of code

````python
# Using debugger called idb 

import ipdb

def mutate(a_list):
    b_list = []
    for i in a_list:
        new_item = i * 2
        ipdb.set_trace()  # Set a breakpoint here
        b_list.append(new_item)
    print(b_list)

# Execute function
mutate([1,2,3,5,8,13])
````
