```python
# Display Banners and VS 


mainBan = '''
||| HIGHER - LOWER |||
||| HIGHER - LOWER |||
||| HIGHER - LOWER |||
'''
rprint(f'[#afff00]{mainBan}')

vsBan = '''
================
VVVVVVVVVVVVVVV
SSSSSSSSSSSSSSS
================
'''
```

Following the solving procedures a mentiond in the course

## Identifying the problem 

```python
# Importing the randomizer
import random as rp 

def format_data(account):
    """Format Panty"""
    account_name = account["name"]
    account_descr = account['description']
    account_country = account['country']
    return f'[#00afff]{account_name} , a {account_descr} from {account_country}'

def check_answers(guess, a_followers, b_followers):
    """Check guess against follower accounts"""
    if a_followers > b_followers:
        return guess == "a"
    else:
        return guess == "b"


console = Console(width=40)
console.print(f'[#afffff on #005f00]{mainBan}', justify="center")

# Keeping the score 
score = 0
game_cont = True
    
# This section needs to keep repeating

while game_cont:
    accA = rp.choice(data)
    accB = rp.choice(data)
    if accA == accB:
        accB = rp.choice(data)
    else:
        rprint(f'Uniquepanty {accA} {accB}')

    rprint(f'Compare A : { format_data(accA) }')
    console.print(f'[#005f00]{vsBan}', justify="center")
    rprint(f'Compare B : { format_data(accB) }')

    # Check if user got it right 
    a_follower_account = accA["follower_count"]
    b_follower_account = accB["follower_count"]
    rprint(f'''
    a_follower_account = {a_follower_account}
    b_follower_account = {b_follower_account}
    ''')

    # Asking guess 
    guess = Prompt.ask('Which Mistress u Want ? A or B', choices = ["a", "b"]).lower()


    ## Checking using function 
    is_correct = check_answers(guess, a_follower_account, b_follower_account)

    # Give feedback of the guesses 
    if is_correct:
        score += 1
        rprint(f'[#00ff00]✅✅✅ Score {score}')
    else:
        game_cont = False
        rprint(f'[#d70087]❌❌❌ Final {score}')
```
