```python
# Menu Dictionary to be used 
MENU = {
    "espresso": {
        "ingredients": {
            "water": 50,
            "coffee": 18,
        },
        "cost": 1.5,
    },
    "latte": {
        "ingredients": {
            "water": 200,
            "milk": 150,
            "coffee": 24,
        },
        "cost": 2.5,
    },
    "cappuccino": {
        "ingredients": {
            "water": 250,
            "milk": 100,
            "coffee": 24,
        },
        "cost": 3.0,
    }
}

resources = {
    "water" : 300, 
    "milk" : 200,
    "coffee" : 100
}
```

```python
# Banners 

shut_down = '''
[#d70000]SHUTTING [#d700d7]DOWN...[#d70000]SHUTTING [#d700d7]DOWN..
[#d70000]SHUTTING [#d700d7]DOWN...[#d70000]SHUTTING [#d700d7]DOWN..
[#d70000]SHUTTING [#d700d7]DOWN...[#d70000]SHUTTING [#d700d7]DOWN..
'''
```

```python
# Important Variables

is_on = True
profit = 0


# Function is resources sufficient 
def is_resources_sufficient(order_ingredients):
    """Return True = Order possible , 
    False = Insufficient Ingredients"""
    for i in order_ingredients:
        if order_ingredients[i] >= resources[i]:
            rprint(f'[#af0087]ğŸ˜¡ no {i}')
            return False
    return True

# Process Coins 
def process_coins():
    """Checking inserted coins"""
    rprint(f'[#af00ff]ğŸª™Give Coin Fukr')
    total = int(Prompt.ask('How Much Qua($0.25): ')) * 0.25
    total += int(Prompt.ask('How Much Dim($0.10): ')) * 0.10
    total += int(Prompt.ask('How Much Nic($0.05): ')) * 0.05
    total += int(Prompt.ask('How Much Pen($0.25): ')) * 0.01
    return total

# Function for transaction 
def is_transaction_sucessful(money_received, drink_cost):
    """Return True if payment accepted ,
    False if its sufficient"""
    if money_received >= drink_cost:
        change = round((money_received - drink_cost),2)
        rprint(f'[#afff5f] Change = {change}')
        global profit
        profit += drink_cost
        return True
    else:
        rprint(f'[#ff005f]NOT ENOUGH COIN')
        return False


while is_on:
    choice = Prompt.ask('''
    ---
    What u want ? 
    es = Espresso - $1.50
    la = latte = $2.50
    ca = cappuccino $3.00
    off = KILL
    report = Print Report
    ---
    ''', choices = ["espresso", "latte", "cappuccino", "off", "report"])
    
    if choice == "off":
        is_on = False
        rprint(shut_down)
        
    elif choice == "report":
        rprint(Panel.fit(f'''[#87ffff]
        Water  : {resources['water']}ml
        Milk   : {resources['milk']}ml
        Coffee : {resources['coffee']}g
        Money  : $ {profit}        
        ''', title="[#d7d700]CofeeMachineReport"))
        is_on = False
        rprint(shut_down)
    else:
        drink = MENU[choice]
        pprint(drink, expand_all = True)
        if is_resources_sufficient(drink["ingredients"]):
            payment = process_coins()
            rprint(Panel.fit(f'Your Payment = {payment}', title="Pussy"))
            is_transaction_sucessful(payment, drink["cost"])
            
    
```
