# Practice pulling data from an API 

1. You will use [JSONPlaceHolder](https://jsonplaceholder.typicode.com/) - get some fake data. Often used when you learning JS and TS 

```python
import requests
import json

def make_request(url):
    # Make a GET request
    response = requests.get(url)

    # The response object has a status_code attribute that gives you the HTTP status code
    if response.status_code == 200:
        # If the request was successful, the response data can be accessed via the .json() method
        data = response.json()

        # Now you can store this data as needed. For example, you might write it to a file:
        with open('data.json', 'w') as file:
            file.write(json.dumps(data))
        return data
    else:
        print(f"Request failed with status code {response.status_code}")
        return None
```

```python
# use function with json placeholder now 
make_request("https://jsonplaceholder.typicode.com/users")
```

```python
# See the data object 
pprint(data)
```

```python
# To see the data you have to get the output into a variable first 

date_fetch1 = make_request("https://jsonplaceholder.typicode.com/users")
pprint(date_fetch1)
rprint(date_fetch1[0])
```

```python
# Running the fetch command with a limit 
data_fetch2_limit = make_request("https://jsonplaceholder.typicode.com/users/2")
pprint(data_fetch2_limit)
inspect(data_fetch2_limit)
```
