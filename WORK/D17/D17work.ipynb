# 156 - Create own Class <a class="anchor" id="s1"></a> [↩️TOC](#toc)

```python
# Creating a class 

class User:
    pass
    
    
# Instantiation of the class 
user_1 = User()

# Adding an attribute 
user_1.id = "001"
user_1.name = "Bina"

rprint(f'''
id   = {user_1.id}
name = {user_1.name}
''')
```

```python
# Using constructor 

class User2:
    def __init__(self):
        rprint('Suck')
        
# Instantiating the class 
user_2 = User2()
user_2.id = "99"
user_2.id = "Nina"

rprint(user_2)
rprint(user_2.id)
```

```python
# adding parameters 
class User3:
    # Constructor
    def __init__(self, user_id, username, follower):
        rprint(f'User3 Class')
        self.id = user_id
        self.username = username # The name should be same as parameter
        self.follower = 0
        

# Instantiate 
user_3 = User3("001", "Fony", 90)

rprint(f'{user_3.id}, {user_3.username}')

# Default can be printed like this
rprint(user_3.follower)
```
