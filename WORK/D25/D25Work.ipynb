# Generate CSV <a class="anchor" id="s1"></a> [↩️TOC](#toc)

1. Generating the csv file with the help of pandas to be used in the course

```python
# Code generated with Mistras

import pandas as pd
import numpy as np

# Generate dummy data
data = {
    'day': pd.date_range(start='2024-01-01', periods=10, freq='D'),
    'temp': np.random.randint(10, 30, size=10),
    'condition': np.random.choice(['Sunny', 'Rainy', 'Cloudy'], size=10)
}

# Create a DataFrame
df = pd.DataFrame(data)

# Save the DataFrame to a CSV file
df.to_csv('we_da.csv', index=False)

print("CSV file 'we_da.csv' has been generated.")
```
