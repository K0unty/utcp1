```python
# Import the random module and rename it to rp for convenience
import random as rp

# Import the print function from the rich module and rename it to rprint for convenience
from rich import print as rprint

# Import the Prompt class from the rich.prompt module
from rich.prompt import Prompt

# Define a list of words to choose from
wor_li = ['ass', 'pussy', 'armpit', 'bbw', "sweaty"]

# Generate a random index to choose a word from the list
rand_index = rp.randint(0, len(wor_li) - 1)

# Choose a random word from the list using the generated index
ran_wo = wor_li[rand_index]

# Print the chosen word (for testing purposes)
rprint(f'Random Chosen word - {ran_wo}')

# Initialize an empty list to keep track of the guessed letters
guessed_letters = []

# Initialize a list of underscores to represent the progress made in guessing the word
word_progress = ['_'] * len(ran_wo)

# Loop until the word has been guessed completely
while '_' in word_progress:
    # Print the current progress made in guessing the word
    rprint(' '.join(word_progress))

    # Prompt the user to enter a letter
    gues1 = Prompt.ask('Enter Letter Bast: ')

    # Convert the guessed letter to lowercase
    gues_lo = gues1.lower()

    # Check if the guessed letter has already been guessed
    if gues_lo in guessed_letters:
        # Print a message indicating that the letter has already been guessed
        rprint(f'You already guessed "{gues_lo}"')

        # Skip to the next iteration of the loop
        continue

    # Add the guessed letter to the list of guessed letters
    guessed_letters.append(gues_lo)

    # Check if the guessed letter is in the word
    if gues_lo in ran_wo:
        # Print a message indicating that the letter is in the word
        rprint(f'✅')

        # Loop through the letters in the word
        for i in range(len(ran_wo)):
            # Check if the current letter is the guessed letter
            if ran_wo[i] == gues_lo:
                # Update the progress made in guessing the word
                word_progress[i] = gues_lo
    else:
        # Print a message indicating that the letter is not in the word
        rprint(f'❌ "{gues_lo}" is not in the word')

# Print a congratulatory message when the word has been guessed correctly
rprint(f'Congratulations! You guessed the word: {ran_wo}')
```

```python
import random
from rich import print as rprint

# Define a list of words to choose from
word_list = ['ass', 'pussy', 'armpit', 'bbw', 'sweaty']

# Choose a random word from the list
chosen_word = random.choice(word_list)

# Initialize the display list with underscores
display = ['_'] * len(chosen_word)

# Set the end_of_game flag to False
end_of_game = False

# Loop until the end_of_game flag is set to True
while not end_of_game:
    # Get a letter guess from the user
    guess = input('Guess a letter: ').lower()

    # Check if the guessed letter is in the chosen word
    for position in range(len(chosen_word)):
        if chosen_word[position] == guess:
            # Update the display list with the guessed letter
            display[position] = guess

    # Print the current state of the display list
    rprint(display)

    # Check if the user has guessed all the letters in the word
    if '_' not in display:
        end_of_game = True
        rprint('You win!')
```
