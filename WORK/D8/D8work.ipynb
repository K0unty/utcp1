## 86 - IC - Prime Number Calculator <a class="anchor" id="s86"></a> [↩️TOC](#toc)

```python
# Prime Number Calculator
def pr_ch(number):
    is_prime = True
    for i in range(2, number):
        if number % i == 0:
            is_prime = False
    if is_prime:
        rprint('✅ Prime')
    else:
        rprint('❌ Not')
    
pr_ch(11)
```

> Find all prime numbers from 1 to 100 from claude

```python
def sieve_of_eratosthenes(n):
    # Create a boolean array "is_prime[0..n]" and initialize
    # all entries it as true. A value in is_prime[i] will
    # finally be false if i is Not a prime, else true.
    is_prime = [True] * (n + 1)
    is_prime[0] = is_prime[1] = False

    for i in range(2, int(n**0.5) + 1):
        if is_prime[i]:
            # Update all multiples of i
            for j in range(i*i, n+1, i):
                is_prime[j] = False

    # Collect all prime numbers
    primes = [i for i in range(2, n+1) if is_prime[i]]
    return primes

def check_primes_up_to(n):
    primes = sieve_of_eratosthenes(n)
    
    print(f"Prime numbers up to {n}:")
    print(primes)
    
    print(f"\nNumber of primes found: {len(primes)}")
    
    # Optionally, you can return the list of primes
    return primes

# Example usage
n = 100
check_primes_up_to(n)
```

## 87-90 - Project Ceasar Cypher  <a class="anchor" id="s879"></a> [↩️TOC](#toc)

This section will have all the lectures from 87 until 90 for building the _Ceasar Cypher_ algorithm

```python
# starting code 
my_ban()
# Automated way of making list of alphabet characters 
import string
alph = list(string.ascii_lowercase)
print(alph)  # Output: Alphabet

# Redoing the list of alphabet to loop over the list 
alphabet = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z','a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']

# Other starter code 
direction = Prompt.ask('''
Type e to - Encrypt 
Type d to - Decrypt
''')
text = Prompt.ask('Msg: ').lower()
shift = int(Prompt.ask('Type Shift Number'))

# start work 
def encrypt(plain_text, shift_amount):
    cipher_text = ""
    for letter in plain_text:
        position = alphabet.index(letter)
        new_position = position + shift_amount
        new_letter = alphabet[new_position]
        cipher_text += new_letter
    rprint(f'EncPussy = {cipher_text}')

# Testing the function

def decrypt(cipher_text, shift_amount):
    plain_text = ""
    for letter in cipher_text:
        position = alphabet.index(letter)
        new_position = position - shift_amount
        plain_text += alphabet[new_position]
    rprint(f'DecAss = {plain_text}')

# now setting up the direction 
if direction == 'e':
    encrypt(plain_text=text, shift_amount=shift)
elif direction == 'd':
    decrypt(cipher_text=text, shift_amount=shift)
```
