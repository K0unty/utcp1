```python
# Copying dict 
dic3 = {
    "Red" : "Mean Stop",
    "green" : "Mean Go",
    "Yellow" : "Mean Ready",
}

# Copy dic3 to dic4 

dic4 = dic3
rprint(dic4)
inspect(dic4)
```

```python
#loop dic
dic4 = {
    "Red" : "Mean Stop",
    "green" : "Mean Go",
    "Yellow" : "Mean Ready",
    "coloArr" : ["pink", "chatreuse", 99, 22]
}
rprint(dic4)

# Loop through 

for i in dic4:
    rprint(i)
    rprint(dic4[i])
    
# Extract the array from the dict and thenprint it 
dic4li = dic4["coloArr"]
rprint(dic4li)

rprint(f'[magenta1] Printing out the array')
for i in dic4li:
    rprint(i)
```
