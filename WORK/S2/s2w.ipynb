## 27 - Day2 - Projcct Tip Calculator <a class="anchor"  id="s27"></a> [↩️TOC](#toc)

````python
# Tip Calculator 
console.rule('27 - Tip Calculator')
billa = float(Prompt.ask('What total amount of bill'))
tipa = float(Prompt.ask('How Much Tip ?'))
peepa = int(Prompt.ask('How Many Kutta'))
rprint(f'''
TT Bill           = {billa}
Tip Given         = {tipa}
Bill with tip     = {billa + (billa * (tipa/100))}
Number of peepole = {peepa}
Bill Per Person   = {round(((billa + (billa *(tipa/100)))/peepa),2)}
''')
````
