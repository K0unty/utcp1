## 34 - EX - BMI 2.0 <a class="anchor"  id="s34"></a> [↩️TOC](#toc)

````python
## console.rule('34 BIM Calculator')
console.rule('34: BMI')
m_ht = Prompt.ask('ht? ')
m_wt = Prompt.ask('wt? ')
m_bmi = float(m_wt) / (float(m_ht) ** 2)
rprint(f'''
Ht = {m_ht} , Wt {m_wt} ,BMI = {int(m_bmi)}kg/m2''')
if int(m_bmi) <= 18.5:
    rprint(f'[deep_sky_blue1]BMI = {m_bmi} UNDERWEIGHT')
elif int(m_bmi) >= 18.5 and int(m_bmi) <=25:
    rprint(f'BMI = {m_bmi} NORMAL')
elif int(m_bmi) >= 25 and int(m_bmi) < 30:
    rprint(f'BMI = {m_bmi} CHUBBY')
elif int(m_bmi) >= 30 and int(m_bmi) < 35:
    rprint(f'[magenta]BMI = {m_bmi} FAT')
else:
    rprint(f'[bold red]BMI = {m_bmi} BUFFALO')
````

## 35 - EX - LeapYear <a class="anchor"  id="s35"></a> [↩️TOC](#toc)

 Description of a leap year 
 1. Year Divisible by 4 = no remainder
 2. Year Evenly divisible by 100 with no remainder 
 3. Year Divisible by 400 no remainder 

````python
# Cheack if the year is aleap year - Taken from codestral
console.rule('35: Leap Year')
ye1 = int(Prompt.ask('Enter year: '))
if ye1 % 400 == 0:
    rprint(f"[dark_olive_green2]{ye1} is a leap year.")
elif ye1 % 100 == 0:
    rprint(f"[orchid]{ye1} is not a leap year.")
elif ye1 % 4 == 0:
    rprint(f"[dark_olive_green2]{ye1} is a leap year.")
else:
    rprint(f"[orchid]{ye1} is not a leap year.")
    
````

## 36 - Multiple If Else Statements <a class="anchor"  id="s36"></a> [↩️TOC](#toc)

````python
bill = 0
rprint(f'''
RollerPussy Rides
''')
ht1 = int(Prompt.ask('What Height: ?'))
if ht1 >= 120:
    rprint(f'OK u RIDE!')
    age1 = int(Prompt.ask('What age u ? :'))
    if age1 < 12:
        bill = 5
        rprint(f'Baccha Price {bill} ')
    elif age1 <= 18:
        bill = 7
        rprint(f'Big Baccha {bill}')
    elif age1 >= 18:
        bill = 12
        rprint(f'Bum {bill}')
    
    want_foto = Prompt.ask('Want Foto?: ', choices=["Y", "N"], default="N")
    if want_foto == "Y":
        bill += 3
    rprint(f'Foto = 3 , Total = {bill}')
else:
    rprint(f'Too TINY')
    
    
````
