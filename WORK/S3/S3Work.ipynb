## 35 - EX - LeapYear <a class="anchor"  id="s35"></a> [↩️TOC](#toc)

 Description of a leap year 
 1. Year Divisible by 4 = no remainder
 2. Year Evenly divisible by 100 with no remainder 
 3. Year Divisible by 400 no remainder 

````python
# Cheack if the year is aleap year - Taken from codestral
console.rule('35: Leap Year')
ye1 = int(Prompt.ask('Enter year: '))
if ye1 % 400 == 0:
    rprint(f"[dark_olive_green2]{ye1} is a leap year.")
elif ye1 % 100 == 0:
    rprint(f"[orchid]{ye1} is not a leap year.")
elif ye1 % 4 == 0:
    rprint(f"[dark_olive_green2]{ye1} is a leap year.")
else:
    rprint(f"[orchid]{ye1} is not a leap year.")
    
````
