Making a rock paper and scissor program

```python
console.rule('49: My Solution')

# Setting up the emoji 
rock_c = "🪨🪨🪨"
paper_c = "📄📄📄"
sci_c = "✂️✂️✂️"

# Your Choice
myChoice = int(Prompt.ask('Choose 1, 2, 3'))
if myChoice == 1:
    rprint(f'Your Chose  - Rock {rock_c}')
elif myChoice == 2:
    rprint(f'Your Chose  - Paper {paper_c}')
else:
    rprint(f'Your Chose  - Scissor {sci_c}')

# Computer Choice 
comp_choice = rp.randint(0,2)
if comp_choice == 1:
    rprint(f'comp_choice  - Rock {rock_c}')
elif comp_choice == 2:
    rprint(f'comp_choice  - Paper {paper_c}')
else:
    rprint(f'comp_choice  - Scissor {sci_c}')

    # Determine the winner
if myChoice == comp_choice:
    rprint('It is a tie!')
elif (myChoice == 'rock' and comp_choice == 'scissors') or \
     (myChoice == 'scissors' and comp_choice == 'paper') or \
     (myChoice == 'paper' and comp_choice == 'rock'):
    rprint('You win!')
else:
    rprint('Computer wins!')
```
