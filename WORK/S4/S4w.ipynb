```python
# Importing libs 
import random as rp 
```

```python
# Randon Number Generation 

import random as rp

rand_1 = rp.randint(2, 1000)
rprint(rand_1)

# Getting random number of 100 digits 
rand_2 = rp.getrandbits(100)
rprint(f'''
[green] Random Numbes of 100 Digits
[magenta] {rand_2}
''')

# Running this print random int in a loop and print out the sequence
for i in range(5):
    rand_3 = rp.getrandbits(100)
    rprint(f'''
    [bright_cyan] Sequence number {i}
    [green] Random Numbes of 100 Digits
    [magenta] {rand_3}
    ''')
```

```python
# creating random python umber 

for i in range(5) :
    random_float = rp.random()
    rprint(random_float)
```

```python
    
# creating random python number for 1 to 5 
for i in range(5) :
    random_float = (rp.random()) * 5
    rprint(random_float)
```

```python
# creating 
for i in range(5) :
    random_float = rp.randint(2,10)
    rprint(random_float)
```
