```python
# Randon Number Generation 

import random as rp

rand_1 = rp.randint(2, 1000)
rprint(rand_1)

# Getting random number of 100 digits 
rand_2 = rp.getrandbits(100)
rprint(f'''
[green] Random Numbes of 100 Digits
[magenta] {rand_2}
''')

# Running this print random int in a loop and print out the sequence
for i in range(5):
    rand_3 = rp.getrandbits(100)
    rprint(f'''
    [brightcyan] Sequence number {i}
    [green] Random Numbes of 100 Digits
    [magenta] {rand_3}
    ''')

    
```
