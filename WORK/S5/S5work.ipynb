## 53 - Average Height Excercise <a class="anchor"  id="s53"></a> [↩️TOC](#toc)

Calculating average height of pussies

```python
# Execrcise

# Getting intput and splitting
console.rule('53: Average Height')
s_h = Prompt.ask('Enter Heights: (any number)')
s_h2 = s_h.split()

# Printing after splitting 
rprint(f'''
Heights Entered  = {s_h2}
''')

# This converts each heing from string to int - now s_h2 will have integers
for n in range(0, len(s_h2)):
    s_h2[n] = int(s_h2[n])

# Set holder variables
tt_h = 0
n_s = 0

# Note at this point the s_h2 has already been converted into integers to you can just access the elements of s_h2
# Calculate the height by looping and holding variable
for h in s_h2:
    tt_h += h
rprint(f'H Total = {tt_h}')

# Get the total number - mimicking the len function
for n in s_h2:
    n_s +=1
rprint(f'N total = {n_s}')

# Now Calculate the average 
av_hn = tt_h / n_s
av_hn_r = round(tt_h / n_s)
rprint(Panel.fit(f'''
Heights Entered = {s_h2}
Average Height = {av_hn}
Average Height Rounded = {av_hn_r}
'''))


    
    
```
