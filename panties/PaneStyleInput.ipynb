```python
def my_st(a):
    console.print(Panel.fit(f'''
    You entered = {a}
    TT ODD = {a % 2 != 0}
    TT Even = {a % 2 == 0}
    ''', title=f"[#87ff00][italic]{a}", border_style="#d700d7"))

my_st(99)
```
