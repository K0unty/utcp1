# Writing Files to disk 

1. You can write files to disk when it is stored in variable 
2. Below we will store the list in a file 

> When session is over the files are lost

```python
import os 

# Store needed text into a variable 
write = '''
my_stuff = ['car', 'money', 'harem']
'''

# Write it to disk 
def WriteFile(var, name):
    '''Function to variable to disk'''
    with open(f'{name}.py', 'w') as f:
        # Write the contents of the variable to the file
        f.write(var)

# Set File Name 

        
# Use function to write file 
WriteFile(write, "panty")

# list file and cat file 
os.system('ls -al *.py')
os.system('cat panty.py -n')
```
